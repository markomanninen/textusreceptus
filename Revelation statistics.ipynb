```python
import strongs
from isopsephy import isopsephy
```

```python
# revelation words in order
words = ' '.join(tr.dataframe[tr.dataframe['orig_book_index'] == "Revelation"].text).split(" ")
# get word data from vocabulary
df = tr.textus_vocabulary.T.ix[words]
# add base word column
df[4] = df[0].map(lambda lemma: ''.join(strongs.find(lemma+'$', 'lemma').word))
# add translation column
df[5] = df[0].map(lambda lemma: ''.join(strongs.find(lemma+'$', 'lemma').translation))
```

```python
# Word, Roman, Lemma, Isopsephy, Root, RootIsopsephy
# adding word isopsephy value to the dataframe
df['root_isopsephy'] = df[4].map(lambda x: isopsephy(x))
```

```python
def primes(n):
    n = int(n)
    divisors = tuple(d for d in range(2, n//2+1) if n % d == 0)
    return tuple(d for d in divisors if all(d % od != 0 for od in divisors if od != d)) if len(divisors) else tuple([n])

df['primes'] = df[2].map(lambda x: primes(x))
df['root_primes'] = df['root_isopsephy'].map(lambda x: primes(x))
```
