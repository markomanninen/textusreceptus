```python
import strongs, math
from isopsephy import isopsephy

def primes(n):
    n = int(n)
    divisors = tuple(d for d in range(2, n//2+1) if n % d == 0)
    return tuple(d for d in divisors if all(d % od != 0 for od in divisors if od != d)) if len(divisors) else tuple([n])

temp_dict = {}

def get_root(lemma):
    if lemma not in temp_dict:
        temp_dict[lemma] = strongs.find(lemma+'$', 'lemma')
    return temp_dict[lemma]

def get_root_word(lemma):
    return ''.join(get_root(lemma).word)

def get_root_translation(lemma):
    return ''.join(get_root(lemma).translation)

sqrt2 = math.sqrt(2)
sqrt3 = math.sqrt(3)
sqrt5 = math.sqrt(5)
sqrt6 = math.sqrt(6)
pi = math.pi
phi = (1+sqrt5)/2

def ratios(x):
    x=int(x)
    return tuple(map(round, [x/sqrt2, x*sqrt2, x/sqrt3, x*sqrt3, x/sqrt5, x*sqrt5, x/sqrt6, x*sqrt6, x/phi, x*phi, x/pi, x*pi]))

def occurrences(word, df):
    return ', '.join(df[df['text'].str.contains(" %s |^%s | %s$" % (word, word, word))].location)
```

```python
# revelation words in order
rev = tr.dataframe[tr.dataframe['orig_book_index'] == "Revelation"].copy()
rev['location'] = rev['orig_chapter'].astype(str) + ":" + rev['orig_verse'].astype(str)
words = ' '.join(rev.text).split(" ")
# get word data from vocabulary
df = tr.textus_vocabulary.T.ix[words]
# filter columns
df = df[df[1].str.contains("^A-|ADV|^G|HEB|^N-|^V-")]
# add base word column
df[3] = df[3].map(lambda roman: roman.replace('w', 'ō').replace('h', 'ē').replace('q', 'th').replace('y', 'ps'))
# add base word column
df['root'] = df[0].map(lambda lemma: get_root_word(lemma))
# add translation column
df['root_translation'] = df[0].map(lambda lemma: get_root_translation(lemma))
# adding word isopsephy value to the dataframe
df['root_isopsephy'] = df['root'].map(lambda x: isopsephy(x))
df['primes'] = df[2].map(lambda x: primes(x))
df['root_primes'] = df['root_isopsephy'].map(lambda x: primes(x))
df['geometry'] = df[2].map(lambda x: ratios(x))
df['root_geometry'] = df['root_isopsephy'].map(lambda x: ratios(x))
df['occurrences'] = df.index.map(lambda x: occurrences(x, rev))
```
