## Data

The Textus Receptus is retrieved from http://unbound.biola.edu/index.cfm?method=downloads.showDownloadMain with complete parsing information for all Greek words; base text is Stephens Textus Receptus (1550), with variants of Scrivener (1894).

File greek_textus_receptus_parsed_utf8.txt is manually pre processed by removing comments on header and saved on data_original/textus_receptus/ directory.

Textusreceptus.py Python module has a function called load_dataframe, which will seek final processed file from data_processed/textus_receptus/ directory. If processed file and dictionary is found, then application uses them, otherwise processing is activated. But even on that case, it should take only 10 seconds to complete first code block from this notebook.

```python
# show an item and its properties by greek word
import json
from isopsephy import to_greek, to_roman
import strongs

def get_vocabulary_word_by_roman(item):
    return get_vocabulary_word(to_greek(item.lower()))

def get_vocabulary_word_by_strongs(item):
    x = tr.textus_vocabulary.T[tr.textus_vocabulary.T[0] == item]
    return get_vocabulary_item(to_greek(x[3][0]), [x[0][0], x[1][0], x[2][0], x[3][0]])

def get_vocabulary_word_by_isopsephy(item):
    result = []
    for x in tr.textus_vocabulary.T[tr.textus_vocabulary.T[2] == str(item)].iterrows():
        result.append(get_vocabulary_item(x[0], [x[1][0], x[1][1], item, to_roman(x[0])]))
    return result

def get_vocabulary_word(item):
    return get_vocabulary_item(item, tr.textus_vocabulary[item])

def get_vocabulary_item(word, item):
    return json.dumps({"Word": word, 
                       "Transliteration": item[3], 
                       "Translation": ''.join(strongs.find(item[0]+'$', 'lemma').translation), 
                       "Isopsephy": item[2], 
                       "Strongs": item[0], 
                       "Morphology": item[1]})

word = 'αββα'

print get_vocabulary_word(word)
print get_vocabulary_word_by_roman("ABBA")
print get_vocabulary_word_by_strongs("G5")
get_vocabulary_word_by_isopsephy(864)
```

```python
# styling the search result
from remarkuple import helper as h
from IPython.display import HTML
from isopsephy import to_roman

"""
def intersection_indexes(l1, l2):
    #return [[i for i, xi in enumerate(l1) if xi == yi] for yi in l2]
    return [range(i,i+len(l2)) for i in xrange(len(l1)-len(l2)+1) if l2 == l1[i:i+len(l2)]]
"""
def list_rows(rows):
    ul = h.ul()
    for i, row in rows.iterrows():
        words = row.text.split(" ")
        for i in row.search:
            for j in i:
                words[j] = str(h.b(words[j], style="color: brown; text-decoration: underline; cursor: pointer", title=to_roman(words[j])))
        reference = h.span(h.b("%s %s:%s" % (row['orig_book_index'], row.orig_chapter, row.orig_verse)))
        ul += h.li(reference, " - ", h.span(' '.join(words)))

    return ul
    
list_rows(rows)
```

```python
from IPython.core.display import HTML
from remarkuple import helper as h, table
from isopsephy import isopsephy, to_roman

tables = ""
for i, row in tr.dataframe[tr.dataframe['total_isopsephy'] == 3617].iterrows():
    tbl = table()
    tbl.addCaption("%s %s:%s %s" % (row.orig_book_index, row.orig_chapter, row.orig_verse, row.text))
    tbl.addHeadRow(h.tr(*map(h.th, ["Strong's", "Transliteration", "Greek", "Isopsephy", "Morphology"])))
    for word in row.text.split():
        item = tr.textus_vocabulary[word]
        tbl.addBodyRow(h.tr(*map(h.td, [item[0], item[3], word, item[2], item[1]])))
    tables = tables + str(tbl)

HTML(tables)
```
